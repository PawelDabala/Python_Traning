```python
frame.loc[1, 'TotalFat'] = np.NaN
frame.loc[1, 'Saturated Fat'] = np.NaN
frame.loc[2, 'Saturated Fat'] = np.NaN
frame.loc[3, 'Category'] = np.NaN
```

```python
frame.fillna(value=0).head()
```

```python
replaceDefinitionTable = {'Category': 'UNKNOWN', 'TotalFat': 0, 'Saturated Fat': 0}
```

```python
replaceDefinitionTable
```

```python
frame.fillna(value=replaceDefinitionTable).head()
```

```python
frame['Category'].fillna(value='UNKNOWN', inplace=True)
```
