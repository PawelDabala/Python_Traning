```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
bar = pd.read_csv('weather_barcelona.csv', index_col='Date')
```

```python
rom = pd.read_csv('weather_rome.csv', index_col='Date')
```

```python
ams = pd.read_csv('weather_amsterdam.csv', index_col='Date')
```

```python
bar.head()
```

```python
temp_max = pd.DataFrame(index=  bar.index)
```

```python
temp_max['Barcelona'] = bar['TempMax']
temp_max['Rome'] = rom['TempMax']
```

```python
temp_max['Amsterdam'] = ams['TempMax']
```

```python
temp_max.head()
```

```python
temp_max.plot()
```

```python
temp_max.plot(figsize=(10,5))
```

```python
temp_max.plot(figsize=(10,5), subplots=True)
```

```python
temp_max.plot(figsize=(10,5), use_index=False)
```

```python
temp_max.plot(figsize=(10,5), title = 'Max temperature by city')
```

```python
temp_max.plot(figsize=(10,5), legend=False)
```

```python
temp_max.plot(figsize=(10,5), logy=True)
```

```python
temp_max[:14].plot(figsize=(10,5), xticks=(range(14)), rot=45 )
```

```python
temp_max[:7].plot(figsize=(10,5), table=True )
```
