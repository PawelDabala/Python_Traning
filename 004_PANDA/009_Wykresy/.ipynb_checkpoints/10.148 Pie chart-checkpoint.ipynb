```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
bar = pd.read_csv('weather_barcelona.csv', index_col='Date')
rom = pd.read_csv('weather_rome.csv', index_col='Date')
ams = pd.read_csv('weather_amsterdam.csv', index_col='Date')
temp_max = pd.DataFrame(index=  bar.index)
```

```python
temp_max['Barcelona'] = bar['TempMax']
temp_max['Rome'] = rom['TempMax']
temp_max['Amsterdam'] = ams['TempMax']
temp_max.head()
```

```python
def GradeDay(row):
    if(row['TempMax'] >= row['AvgTempMax']):
        return 'PLUS'
    else:
        return 'MINUS'
    
```

```python
rom['GradeDay'] = rom.apply(GradeDay, axis=1)
bar['GradeDay'] = bar.apply(GradeDay, axis=1)
ams['GradeDay'] = ams.apply(GradeDay, axis=1)
```

```python
nasa = pd.read_csv("nasa_facebook_statuses.csv",
 usecols=["status_published","num_reactions",
 "num_comments","num_shares","num_likes"])
```

```python
#Convert column type to date time
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
```

```python
#Select observations between two datetimes - May 2015
filter =(nasa['status_published'] >= '2016-3-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]
```

```python
#Add column with day only
nasa['month'] = nasa.apply(lambda row: row["status_published"].month,axis=1)
```

```python
#Group all columns by sum
nasa_by_month = nasa.groupby(by='month').sum()
```

```python
nasa_by_month.head()
```

```python
nasa_by_month['num_likes'].plot(kind='pie') 
```

```python
nasa_by_month['num_likes'].plot(kind='pie', figsize=(5,5)) 
```

```python
nasa_by_month['num_likes'].plot(kind='pie', figsize=(5,5), title='NASA Facebook activity', labels=['marzec', 'kwiecień', 'maj']) 
```

```python
nasa_by_month.plot(kind='pie', figsize=(10,10), title='NASA Facebook activity', labels=['marzec', 'kwiecień', 'maj'], layout=(2,2), subplots=True )
```

```python
nasa_by_month.plot(kind='pie', figsize=(10,10), title='NASA Facebook activity', labels=['marzec', 'kwiecień', 'maj'], layout=(2,2), subplots=True, autopct='%.0f%%')
```

```python
nasa_by_month.plot(kind='pie', figsize=(10,10), title='NASA Facebook activity', labels=['marzec', 'kwiecień', 'maj'], layout=(2,2), subplots=True, autopct='%.0f%%', textprops=dict(color='w'))
```
