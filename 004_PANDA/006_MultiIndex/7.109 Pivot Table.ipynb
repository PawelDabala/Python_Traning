```python
df = pd.read_csv('marathon_results_2016.csv.', index_col='Bib', usecols=[ 'Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
 
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
 
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
 
df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
```

```python
df.pivot_table(index='Age', columns='M/F', values='TotalSeconds', aggfunc='mean')
```

```python
df.pivot_table(index='Age', columns='M/F', values=['TotalSeconds', 'HalfSeconds'], aggfunc='mean')
```

```python
df.pivot_table(index='Age', columns='M/F', values=['TotalSeconds', 'HalfSeconds'], aggfunc='mean').swaplevel(axis='columns').sort_index(axis='columns')
```

```python
df.pivot_table(index='Age', columns='M/F', values='TotalSeconds', aggfunc='count' )
```

```python
df.pivot_table(index='Age', values='TotalSeconds', aggfunc='count' )
```

```python
df.pivot_table(index='Country', values='TotalSeconds', aggfunc='count')
```

```python
df.pivot_table(index='Country', values='TotalSeconds', aggfunc='count').sort_values(by='TotalSeconds', ascending=False)
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds', aggfunc='mean')
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds', aggfunc='mean').sort_values(by='TotalSeconds', ascending=False)
```
