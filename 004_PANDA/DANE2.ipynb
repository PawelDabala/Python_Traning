```python
plan2.Copy = plan2.Copy.astype('category')
plan2.Product = plan2.Product.astype('category')
plan2.Ch = plan2.Ch.astype('category')
plan2['TV station'] = plan2['TV station'].astype('category')
plan2.Programme = plan2.Programme.astype('category')
plan2.NrZlecenia = plan2.NrZlecenia.astype('category')
```

```python
columns = ['Copy', 'Product', 'Ch', 'TV station', 'Programme', 'NrZlecenia']
for column in columns:
    plan2[column] = plan2[column].astype('category')
```
