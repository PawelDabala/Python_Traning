```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
import random as rd
from datetime import datetime, time, timedelta
```

```python
plan = pd.read_excel(r"C:\002 PROGRAMOWANIE\015_PythonTraning\004_PANDA\plik kampanii Skoda Octavia.xlsx", sheet_name='PLAN')
plan.columns = plan.iloc[3].values
```

```python
plan.drop(axis='rows', labels=[0, 1, 2, 3], inplace=True)
```

```python
plan.head()
```

```python
plan.set_index('row_nr', inplace=True)
```

```python
plan.columns
```

```python
plan = plan[['Data', 'Copy', 'Product', 'Time', 'Ch', 'TV station', 'Programme', 'NrZlecenia']]
```

```python
columns = ['Copy', 'Product', 'Ch', 'TV station', 'Programme', 'NrZlecenia']
for column in columns:
    plan[column] = plan[column].astype('category')
```

```python
isTime = lambda row: True if isinstance(row["Time"], time) else False
plan["IsTime"] = plan.apply(isTime, axis=1)
```

```python
plan[plan.IsTime == False]
```

```python
addOneDay = (
            lambda row: row.Data + pd.Timedelta(days=1)
            if row["IsTime"] is False
            else row.Data
        )
plan["Data"] = plan.apply(addOneDay, axis=1)
```

```python
conToTime = (
            lambda row: row["Time"].time() if row["IsTime"] is False else row["Time"]
        )
plan["Time"] = plan.apply(conToTime, axis=1)
```
